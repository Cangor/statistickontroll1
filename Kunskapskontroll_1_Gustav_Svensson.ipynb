Del 2 Statistisk dataanalys i Python

- Jag vill förstå vad som påverkar bilpriser för att billigare kunna köpa in bilar för återförsäljning.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Läs in CSV-filen
df = pd.read_csv("bilpriser/bilpriser/car_price_dataset.csv", sep=';')

# Scatterplot för ålder
df["Car_Age"] = 2025 - df["Year"]

plt.figure(figsize=(8, 5))
plt.scatter(df["Car_Age"], df["Price"], alpha=0.5)
plt.title("Samband mellan bilens ålder och pris")
plt.xlabel("Bilens ålder (år)")
plt.ylabel("Pris (kr)")
plt.grid(True)
plt.show()

# Stapeldiagram för märke
brand_avg = df.groupby("Brand")["Price"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(brand_avg.index, brand_avg.values)
plt.title("Genomsnittligt pris per bilmärke")
plt.xlabel("Bilmärke")
plt.ylabel("Pris (kr)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatterplot för milage
plt.figure(figsize=(8,5))
plt.scatter(df["Price"], df["Mileage"], alpha=0.5)
plt.title("Samband mellan bilens milage och pris")
plt.xlabel("Bilens Milage")
plt.ylabel("Pris (kr)")
plt.grid(True)
plt.show()
```

Sammanfattning:

- Nyare bilar är oftast dyrare än äldre bilar.
- Alla bilmärken har i genomsnitt väldigt likt pris.
- Ju längre bilen kört desto billigare är den.
- Denna information kan hjälpa om man vill starta ett bilföretag och veta vilka bilar som är mest lönsamma att köpa och sälja. Med nuvarande data så verker det vara mycket viktigare att sikta på bil ålder samt milage och inte märket.
